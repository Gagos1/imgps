Import Library CV2,Numpy,MathPlotLib

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
from matplotlib import pyplot as plt
```

Membaca Gambar

```python
car = cv2.imread ("/content/drive/MyDrive/Colab Notebooks/car.jpg")
lumba = cv2.imread ("/content/drive/MyDrive/Colab Notebooks/lumba.jpg")
burung = cv2.imread ("/content/drive/MyDrive/Colab Notebooks/burung.jpg")
```

Menampilkan Gambar

Menampilkan Dimensi Gambar

```python
print(car.shape)
print(lumba.shape)
print(burung.shape)
```

Konversi Warna

```python
burungabu = cv2.cvtColor(burung,cv2.COLOR_BGR2GRAY)
burungrgb = cv2.cvtColor(burung,cv2.COLOR_BGR2RGB)
burunghsv = cv2.cvtColor(burung,cv2.COLOR_BGR2HSV)
```

Menampilkan hasil gambar konversi

```python
print("Burung Abu")
cv2_imshow(burungabu)
print("Burung RGB")
cv2_imshow(burungrgb)
print("Burung HSV")
cv2_imshow(burunghsv)
```

Memisahkan Channel BGR

```python
B,G,R = cv2.split(burung)
print("Blue")
cv2_imshow(B)
print("Green")
cv2_imshow(G)
print("Red")
cv2_imshow(R)

gabung = cv2.merge([B,G,R])
print("Gabung")
cv2_imshow(gabung)
```

```python
H,S,V = cv2.split(burunghsv)
print("Hui")
cv2_imshow(H)
print("Saturation")
cv2_imshow(S)
print("Variable")
cv2_imshow(V)

gabung1 = cv2.merge ([H,S,V])
print("Gabung")
cv2_imshow(gabung1)
```

Menampilkan Math Plot Lib

```python
plt.figure(figsize=[20,20])
plt.subplot(321)
plt.imshow(B)
plt.subplot(322)
plt.imshow(G)
plt.subplot(323)
plt.imshow(R)
plt.subplot(324)
plt.imshow(gabung)
plt.subplot(325)
plt.imshow(burungrgb)
plt.subplot(326)
plt.imshow(burunghsv)
plt.show()
```
